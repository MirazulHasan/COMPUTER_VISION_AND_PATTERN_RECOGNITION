```python
from IPython import display
import matplotlib.pyplot as plt
import random
import math
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
X1 = df["X1"].values
X2 = df["X2"].values
Y = df["Y"].values

for i in range(len(Y)):
    if Y[i] == 0:
      plt.plot(X1[i], X2[i], "bo")  # Blue for class 0
    else:
      plt.plot(X1[i], X2[i], "ro")  # Red for class 1
```

```python
W = [random.uniform(-1, 1), random.uniform(-1, 1)]
b = random.uniform(-1, 1)

lr = 0.01
theta = 0.5

epoch = 0
count = 0
loss = []

while count < len(Y):
  total_error = 0

  for i in range(len(Y)):

    ws = W[0]*X1[i] + W[1]*X2[i] + b
    pred = 0

    if ws > theta:
      pred = 1

    error = Y[i] - pred

    if error != 0:
      count = 0
      total_error += math.sqrt(error**2)
      W[0] += lr*error*X1[i]
      W[1] += lr*error*X2[i]
      b += lr*error
    else:
      count += 1

  loss.append(1/len(X1)*total_error)
  epoch +=1

plt.figure(figsize=(20,10))
display.clear_output(wait=True)
plt.subplot(1,2,1)

for i in range(len(Y)):

  if Y[i] == 0:
    plt.plot(X1[i], X2[i], "bo")

  else:
    plt.plot(X1[i], X2[i], "ro")
    xx = plt.xlim()
    xint = (theta-b) / W[0]
    yint = (theta-b) / W[1]
    slope = -yint/xint
    yy = []

    for x in xx:
      yy.append(slope*x+yint)

    plt.plot(xx,yy)
    plt.xlim([-10,110])
    plt.ylim([-10,110])

plt.subplot(1,2,2)
plt.plot(loss)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
